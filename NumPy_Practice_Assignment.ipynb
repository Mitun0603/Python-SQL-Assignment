## Q1. Create a NumPy Array

```

import numpy as np

# Given list
data = [1, 2, 3, 4, 5]

# Create numpy array
arr = np.array(data)

# Print array and its type
print("Array:", arr)
print("Type:", type(arr))
```

## Q2. Array Slicing

```

import numpy as np

# Create array
arr = np.array([10, 20, 30, 40, 50, 60])

# Slice the array
sliced = arr[1:4]

print("Sliced array:", sliced)
```

## Q3. Reshape an Array

```

import numpy as np

# Original array
arr = np.array([1, 2, 3, 4, 5, 6])

# Reshape into 2x3
reshaped = arr.reshape(2, 3)

print("Reshaped array:\n", reshaped)
```

## Q4. Basic Statistics

```

import numpy as np

scores = np.array([80, 85, 90, 95, 100])

# Compute statistics
mean_val = np.mean(scores)
median_val = np.median(scores)
std_val = np.std(scores)

print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)
```

## Q5. Conditional Selection

```

import numpy as np

ages = np.array([12, 18, 25, 17, 30, 15])

# Select ages >= 18
adults = ages[ages >= 18]

print("Ages >= 18:", adults)
```

## Q6. Element-wise Operations

```

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise operations
mul = a * b
add = a + b
sub = a - b

print("Element-wise multiplication:", mul)
print("Element-wise addition:", add)
print("Element-wise subtraction:", sub)
```

## Q7. Matrix Multiplication

```

import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])

# Matrix multiplication
result = np.dot(A, B)

print("Matrix multiplication result:\n", result)
```

## Q8. Broadcasting

```

import numpy as np

A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([10, 20, 30])

# Add using broadcasting
result = A + B

print("Result after broadcasting:\n", result)
```

## Q9. Advanced Indexing

```

import numpy as np

arr = np.array([[5, 10, 15],
                [20, 25, 30],
                [35, 40, 45]])

# Extract specific elements using advanced indexing
result = arr[[0, 1, 2], [1, 2, 0]]

print("Extracted elements:", result)
```

## Q10. Find Most Frequent Value

```

import numpy as np

arr = np.array([1, 3, 2, 1, 4, 1, 3, 3, 2, 2])

# Find most frequent element
(unique, counts) = np.unique(arr, return_counts=True)
most_frequent = unique[np.argmax(counts)]

print("Most frequent element:", most_frequent)
```
